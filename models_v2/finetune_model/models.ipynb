```python
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv("final_data.csv")
# 1. Chia thành Train và tập còn lại (Val + Test) theo tỷ lệ 80/20
# Tham số stratify=df['label_id'] cực kỳ quan trọng ở đây
df_train, df_temp = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42, 
    shuffle=True, 
    stratify=df['label_id'] 
)

# 2. Chia tập còn lại thành Validation và Test (50/50 của 20% là mỗi cái 10%)
df_val, df_test = train_test_split(
    df_temp, 
    test_size=0.5, 
    random_state=42, 
    shuffle=True, 
    stratify=df_temp['label_id']
)

print(f"Train: {len(df_train)} - Val: {len(df_val)} - Test: {len(df_test)}")
```

```python
print("Phân bố tập train:", df_train['label_id'].value_counts())
print("Phân bố tập val:", df_val['label_id'].value_counts())
print("Phân bố tập test:", df_test['label_id'].value_counts())
```
